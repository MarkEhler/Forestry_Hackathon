```python
import pandas as pd
from datetime import datetime
import json
import numpy as np
import pytz
import tensorflow as tf
from tensorflow import keras
import requests
```

```python
#YYYYMMDDHH
init_time=2020101500


request=f"http://api.northviewweather.com/api/v1/resources/forecasts/timeseries?access=77d24c44bdba43c683054b7f983a1652&variables=apcp,t2m,asnow,icethick&region=Towns&model=mesoscale&loc=Barton&init_time={init_time}&statistic=mean&tz=local"

response = requests.get(request)

response = response.json()
```

```python
response
```

```python
precip=response['LOCATIONS']['Barton']['apcp']['mean']
time=response['LOCATIONS']['Barton']['valid_time']
temp=response['LOCATIONS']['Barton']['t2m']['mean']
snow=response['LOCATIONS']['Barton']['asnow']['mean']
ice=response['LOCATIONS']['Barton']['FRAMthick']['mean']
    
time2=[]
for t in range(0, len(time)):
    time2.append((datetime.strptime(time[t], "%Y-%m-%d %H:%M:%S")).astimezone(pytz.timezone('US/Eastern')))
    
data=pd.DataFrame({'Time':time2,'Temp':temp,'Precip':precip,'Snow':snow,'Ice':ice})
```
